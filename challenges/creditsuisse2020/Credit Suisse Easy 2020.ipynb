 <b>Description  </b>   
Emilia, a quantitative researcher, predicts how the closing share price of a stock moves over time.  

She wants to find the maximum possible profit of a stock over a given period of time, using only one buy and one sell operation according to a given sequence of predicted share prices.
<br>

<b> Constraints </b>  
Short selling is not allowed.  
All of the predicted share prices are positive integers.

```python
def main():
    line = [8, 100, 10, 12, 5, 6, 14, 5, 6]
    numOfPredictedDay = int(line[0])
    predictedSharePrices = list(map(int, line[1:]))

    answer = findMaxProfit(numOfPredictedDay, predictedSharePrices)
    print(answer)
   
    
def findMaxProfit(numOfPredictedDay, predictedSharePrices):
    initial_max = predictedSharePrices[-1]
    output = -1
    print("init max: " ,initial_max)
    for i in range(numOfPredictedDay-1):
        print("pos i ", i)
        temp = predictedSharePrices[-(i+1)]
        print("pos i val ", temp)
        if(temp > initial_max):
            initial_max = temp
        elif((initial_max - temp) > output):
            output = initial_max - temp
            print("out ",output)
        
    return output
        
        
    
        
main()
```

### Question 3
 
<b>Description  </b>  
John has recently started stock trading. He has predicted share prices for a particular company, over the next N days. John wants to analyze this data to build a model which will predict the best day to buy and sell the shares to achieve a specific profit. If there are multiple approaches of buying and selling shares to achieve this profit, John would like to know which of these will achieve the profit the earliest.  

<b>Constraints  </b>  
Only 1 share can be bought.  
Short selling is not allowed.  
1 ≤ N ≤ 100000   
1 ≤  D ≤ 10  
1 ≤ Ni, Di ≤ 1000000

```python
def find_min_days(prices, profit):
    return_list = []
    return_string = ""
    for w in profit:
        const = w
        check = []
        minday = -1
        pair = []

        newt= [x for x in prices if x - const in prices]
        newb= [x for x in prices if x + const in prices]
        #print(newt, "h ,l ", newb)

        indicesb = [index for index, element in enumerate(prices) if element in newb]
        indicest = [index for index, element in enumerate(prices) if element in newt]
        #print(indicest, "h , l ", indicesb)

        for x in indicest:
            temp = [y for y in indicesb if x > y] # eliminates values that are in a days after the greater values
            temp.reverse()
            #print(temp)
            for z in temp:
                if(prices[x] - prices[z] == const):
                    check.append([z+1, x+1])
                    #print(check)
                    break

        for x in check:
            if check == []:
                pair.append(-1)
            elif(minday == -1 or (x[1] < minday)):
                minday = x[1]
                pair = x

        return_list.append(pair)
        
    for x in return_list:
        if x != []:
            return_string += "{} {},".format(x[0], x[1])
        else:
            return_string += "-1,"
    
    return_string = return_string[:-1]
    return return_string
    
list = [1,3] # input prices from test suite
const = [5,2] # input buy targets from test suite
answer = find_min_days(list, const)

print(answer)
```
